```python
def get_izos(soup, izo, red_izo):
    tables = soup.find_all('table')
    for table in tables:
        table_body = table.find('tbody')
        tr_rows = table_body.find_all('tr')
        for tr_row in tr_rows:
            td_rows = tr_row.find_all('td')
            for td_row in td_rows:
                try:
                    if 'Gymnázium' in td_row.find_next('td').find_next('td').text:
                        red_izo.append(td_row.text)
                        izo.append(td_row.parent.find_next('tr').find_all('td')[0].text.replace('\xa0\xa0\xa0\xa0', ''))
                    elif 'gymnázium' in td_row.find_next('td').find_next('td').text:
                        red_izo.append(td_row.text)
                        izo.append(td_row.parent.find_next('tr').find_all('td')[0].text.replace('\xa0\xa0\xa0\xa0', ''))
                    elif 'Gymnásium' in td_row.find_next('td').find_next('td').text:
                        red_izo.append(td_row.text)
                        izo.append(td_row.parent.find_next('tr').find_all('td')[0].text.replace('\xa0\xa0\xa0\xa0', ''))
                    elif 'gymnásium' in td_row.find_next('td').find_next('td').text:
                        red_izo.append(td_row.text)
                        izo.append(td_row.parent.find_next('tr').find_all('td')[0].text.replace('\xa0\xa0\xa0\xa0', ''))
                    elif 'GYMNÁZIUM' in td_row.find_next('td').find_next('td').text:
                        red_izo.append(td_row.text)
                        izo.append(td_row.parent.find_next('tr').find_all('td')[0].text.replace('\xa0\xa0\xa0\xa0', ''))
                    elif 'GYMNÁSIUM' in td_row.find_next('td').find_next('td').text:
                        red_izo.append(td_row.text)
                        izo.append(td_row.parent.find_next('tr').find_all('td')[0].text.replace('\xa0\xa0\xa0\xa0', ''))
                except:
                    print('konec')
```
