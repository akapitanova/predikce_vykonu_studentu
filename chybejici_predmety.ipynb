```python
# studium_sub a student_sub

# nahrazeni datumu zahajeni studia a roku maturity, jejich rozdilem
# pokud jedna z hodnot chybi je hodnota nahrazena prumernym rozdilem
studium_sub = studium_dim[['studium_id', 'datum_zahajeni', 'rok_maturity']]
studium_sub['maturita_nastup_rozdil'] = studium_sub['datum_zahajeni'] - studium_sub['rok_maturity']
studium_sub.loc[studium_sub['maturita_nastup_rozdil'] > 1000, 'maturita_nastup_rozdil'] = math.nan
studium_sub.loc[studium_sub['maturita_nastup_rozdil'] < 0, 'maturita_nastup_rozdil'] = math.nan
rozdil_mean = round(studium_sub['maturita_nastup_rozdil'].mean())
studium_sub['maturita_nastup_rozdil'] = studium_sub['maturita_nastup_rozdil'].fillna(rozdil_mean)
studium_sub = studium_sub.drop(columns=['datum_zahajeni', 'rok_maturity'])

# nahrazeni sloupce statni_prislusnost_nazev, novym sloupcem je_cech (1 - pokud ano, 0 - pokud ne)
student_sub = student_dim[['studium_id', 'pohlavi_kod', 'statni_prislusnost_nazev']]
student_sub['je_cech'] = math.nan
student_sub.loc[student_sub.statni_prislusnost_nazev == 'Česko', 'je_cech'] = 1
student_sub.loc[student_sub.statni_prislusnost_nazev != 'Česko', 'je_cech'] = 0
student_sub = student_sub.drop(columns='statni_prislusnost_nazev')
```
