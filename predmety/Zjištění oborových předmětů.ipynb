## Všechny předměty

```python
column_names = ["role", "kod_predmetu", "nazev_predmetu", "rozsah_vyuky", "zakonceni", "semestr", "pocet_kreditu", "vyucujici", "katedra"]
courses_bak_spec = pd.DataFrame(columns = column_names)
courses_mag_spec = pd.DataFrame(columns = column_names)
```

## Bakalář

```python
courses_bak = pd.DataFrame(columns = column_names)

url_bak = "http://bk.fit.cvut.cz/cz/pruchody/pr1241646176405.html"
tables = pd.read_html(url_bak)
tables_nmb = len(tables)

sem = 1
for table_nmb in range(2, tables_nmb - 1, 2):
    tmp = tables[table_nmb].copy(deep=True)
    tmp = tmp.rename(columns={0: "role", 1: "kod_predmetu", 2: "nazev_predmetu", 3: "rozsah_vyuky", 4: "zakonceni", 5: "semestr", 6: "pocet_kreditu", 7: "vyucujici", 8: "katedra"})
    tmp["doporuceny_semestr"] = sem
    courses_bak = courses_bak.append(tmp, ignore_index = True)
    sem += 1

courses_bak = courses_bak.dropna(subset = ["kod_predmetu"])
courses_bak = courses_bak[(~courses_bak['kod_predmetu'].str.contains("Kód předmětu", case=False))]
```

### Oborové

```python
column_names = ["role", "kod_predmetu", "nazev_predmetu", "rozsah_vyuky", "zakonceni", "semestr", "pocet_kreditu", "vyucujici", "katedra", "doporuceny_semestr"]

urls = {"ism":"http://bk.fit.cvut.cz/cz/pruchody/pr620998552005.html", 
        "ti":"http://bk.fit.cvut.cz/cz/pruchody/pr620998470105.html",
        "pi":"http://bk.fit.cvut.cz/cz/pruchody/pr620998511805.html",
        "bit":"http://bk.fit.cvut.cz/cz/pruchody/pr620760770705.html",
        "wsi_pg":"http://bk.fit.cvut.cz/cz/pruchody/pr620998649405.html",
        "wsi_si":"http://bk.fit.cvut.cz/cz/pruchody/pr620998618205.html",
        "wsi_wi":"http://bk.fit.cvut.cz/cz/pruchody/pr620998637105.html",
        "zi_1517":"http://bk.fit.cvut.cz/cz/pruchody/pr620998595905.html",
        "zi_1820":"http://bk.fit.cvut.cz/cz/pruchody/pr938373126205.html"
       }

ism = pd.DataFrame(columns = column_names)
ti = pd.DataFrame(columns = column_names)
pi = pd.DataFrame(columns = column_names)  
bit = pd.DataFrame(columns = column_names)
wsi_pg = pd.DataFrame(columns = column_names)
wsi_si = pd.DataFrame(columns = column_names)
wsi_wi = pd.DataFrame(columns = column_names)
zi_1517 = pd.DataFrame(columns = column_names)
zi_1820 = pd.DataFrame(columns = column_names)

specs = {"ism":ism, "ti":ti, "pi":pi, "bit":bit,
        "wsi_pg":wsi_pg, "wsi_si":wsi_si, "wsi_wi":wsi_wi,
        "zi_1517":zi_1517, "zi_1820":zi_1820}
```

```python
for spec in specs:
    tables = pd.read_html(urls[spec])
    tables_nmb = len(tables)

    sem = 1
    for table_nmb in range(2, tables_nmb - 1, 2):
        tmp = tables[table_nmb].copy(deep=True)
        tmp = tmp.rename(columns={0: "role", 1: "kod_predmetu", 2: "nazev_predmetu", 3: "rozsah_vyuky", 4: "zakonceni", 5: "semestr", 6: "pocet_kreditu", 7: "vyucujici", 8: "katedra"})
        
        courses_bak_spec = pd.concat([courses_bak_spec,tmp]).drop_duplicates().reset_index(drop=True)
        
        tmp["doporuceny_semestr"] = sem
        specs[spec] = specs[spec].append(tmp, ignore_index = True)
        sem += 1
    
    specs[spec] = specs[spec].dropna(subset = ["kod_predmetu"])
    specs[spec] = specs[spec][(~specs[spec]['kod_predmetu'].str.contains("Kód předmětu", case=False))]
    specs[spec].to_csv("{sp:}_bak.csv".format(sp = spec), index=False)
    
    courses_bak_spec = courses_bak_spec.dropna(subset = ["kod_predmetu"])
    courses_bak_spec = courses_bak_spec[(~courses_bak_spec['kod_predmetu'].str.contains("Kód předmětu", case=False))]
```
